```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 진에어 기내식 : 웹 스크래핑 

jinair = 'https://www.jinair.com/addService/preordermeal/all'
driver = webdriver.Chrome()
driver.get(jinair)

WebDriverWait(driver, 15).until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, "#container > div > div.contents-wrapper.extra-service-airplane-food")
    )
)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

container = soup.select_one("#container > div > div.contents-wrapper.extra-service-airplane-food")

items = container.find_all("div", class_="desc-wrap")

jinair_menu = []

for item in items:
    dic = {}
    dic['메뉴 이름'] = item.find('strong').get_text().strip()
    dic['메뉴 가격'] = item.find("p", class_="price").get_text().strip()
    dic['기내식이 제공되는 항공 노선'] = item.find("p", class_="imp").get_text().strip()
    jinair_menu.append(dic)
```

```python
# 제주항공 기내식 : url 제공

jejuair = 'https://www.jejuair.net/ko/additionalService/service/preorderedMeal.do'
driver = webdriver.Chrome()
driver.get(jejuair)

WebDriverWait(driver, 15).until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, "#main > div:nth-child(3) > div.section-wrap > div.list-product.show")
    )
)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

items = soup.find_all("div", class_="list-product__info")
items
jejuair_menu = []

for item in items:
    dic = {}
    dic['메뉴 이름'] = item.find('div',class_='list-product__title').text
    dic['메뉴 가격'] = item.find('span',class_='cost').text.strip()
    jejuair_menu.append(dic)
```
